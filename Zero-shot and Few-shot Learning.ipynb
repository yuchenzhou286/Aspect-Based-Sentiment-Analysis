## 1. Zero-shot Learning

We first looked into a model regarding zero-shot classification from : https://huggingface.co/facebook/bart-large-mnli?candidateLabels=mobile%2C+website%2C+billing%2C+account+access&multiClass=false&text=Last+week+I+upgraded+my+iOS+version+and+ever+since+then+my+phone+has+been+overheating+whenever+I+use+your+app

We used the zero-shot classification pipeline.


We basically labeled the quotations as:
 "'positive', 'negative', 'neutral'" and then the model returns a probability of each label of the whole sentence. 




```python
!pip install datasets evaluate transformers[sentencepiece]
```

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
quotes = reviews['quote'].to_list()
#quotes
```

```python
sequence_to_classify = quotes
candidate_labels = ['positive', 'negative', 'neutral']
pred_result = classifier(sequence_to_classify, candidate_labels)
pred_result
```

```python
# wrap the predicted result 
values = []
for key in pred_result:
  value = [key['sequence'],key['labels'][0]]
  values.append(value)

import pandas as pd

df = pd.DataFrame(values, columns=['quotes', 'sentiment'])
df 
# df is the dataframe which corresponds to the model prediction
```

```python
data = []
for index, values in reviews.iterrows():
  if (pd.isna(values['entity 2']) or pd.isnull(values['entity 2'])) and (pd.isna(values['entity 3']) or pd.isnull(values['entity 3'])):
    value = [values['quote'],values['sentiment 1']]
    data.append(value)


df2 = pd.DataFrame(data, columns=['quote_raw', 'sentiment_manual'])
df2
# df2 is the dataframe that corresponds to the labeled results in the dataset
```

```python
# mapping the numbers into the actual words that represent sentiments
score = {1.0:'negative',
         2.0:'neutral',
         3.0:'positive'}
df2["sentiment"] = df2["sentiment_manual"].map(score)
df2.head()
```

### calculate the model accuracy by comparing with our manually labeled quotation.

```python
# see the prediction accuracy of zero shot learning.
# In this step we only compare quotes that contains only one entity since zero shot learning does not do aspect extraction. 
# So it does not make sense to include quotes that contain multiple entities.

i = 0
for index1, row1 in df2.iterrows():
  for index2,row2 in df.iterrows():
    if row1['quote_raw'] == row2['quotes'] and row1['sentiment'] == row2['sentiment']:
      i+=1

# print the accuracy
print(i/df2.shape[0])
```

### Takeaway from Zero-Shot Learning
Seeing the low model accuracy, I pulled out the quotes that have been labeled 'neutral' in sentiment. 

Since there is a relatively very small portion of quotes with a 'neutral' sentiment, we probably need to 1)enlarge the dataset by manually labling more; 2)change to another model that works well when there are two few instances.

### 2. Fine tuning the SetFit model

```python
#set the seed 
def hp_space(trial):  # Training parameters
    return {
        "learning_rate": trial.suggest_float("learning_rate", 1e-6, 1e-4, log=True),
        "num_epochs": trial.suggest_int("num_epochs", 1,  5),
        "batch_size": trial.suggest_categorical("batch_size", [4, 8, 16, 32, 64]),
        #"seed": trial.suggest_int("seed", 1, 40),
        "num_iterations": trial.suggest_categorical("num_iterations", [5, 10, 20]),
        "max_iter": trial.suggest_int("max_iter", 50, 300),
        "solver": trial.suggest_categorical("solver", ["newton-cg", "lbfgs", "liblinear"]),        
    }
```

```python
def model_init(params):
    params = params or {}
    max_iter = params.get("max_iter", 100)
    solver = params.get("solver", "liblinear")
    params = {
        "head_params": {
            "max_iter": max_iter,
            "solver": solver,
        }
    }
    return SetFitModel.from_pretrained("sentence-transformers/paraphrase-albert-small-v2", **params)
```
